```python
import pandas as pd
import numpy as np
import ast
```

```python
df=pd.read_csv('0.movies_dataset.csv')
```

```python
#Vemos que hay en el dataset
display(df.head())
```

```python
#Creamos un nuevo dataset quitando columnas innecesarias
df1= df.drop(['video', 'imdb_id','tagline', 'adult', 'original_title', 'poster_path', 'homepage'], axis=1)
df1
```

```python
#Creamos una funcion para desanidar la informacion de la columna 'belongs_to_collection'
#Borramos la columna original y creamos 2 nuevas con la info necesaria

def extract_value(x, key):
    if isinstance(x, str):
        try:
            # Intenta convertir la cadena a un diccionario
            x = ast.literal_eval(x)
        except:
            # Si falla, retorna None
            return None

    if isinstance(x, dict):
        return x.get(key)
    else:
        return None

df1['id_collection'] = df1['belongs_to_collection'].apply(lambda x: extract_value(x, 'id'))
df1['name_collection'] = df1['belongs_to_collection'].apply(lambda x: extract_value(x, 'name'))

df1 = df1.drop('belongs_to_collection', axis=1)
```

```python
#A pesar de que desanidamos la columna 'belongs_to_collection', la informacion de la misma no es relevante para nuestro objetivo final
#A continucion las eliminamos 

df1= df1.drop(['id_collection', 'name_collection'], axis=1)
```

```python
#Creamos otra funcion para desanidar las columnas 'production_companies', 'genres', 'production_countries' y 'spoken_languages'

def extract_value1(x, key):
    if isinstance(x, str):
        try:
            x = ast.literal_eval(x)
        except:
            return None
    
    if isinstance(x, list):
        values = [item.get(key) for item in x if isinstance(item, dict)]
        if values:
            return ', '.join(map(str, values)) 
        else:
            return None
    
    elif isinstance(x, dict):
        value = x.get(key)
        return str(value)  
    
    else:
        return None
    
```

```python
#production_companies
df1['production_companies'] = df1['production_companies'].apply(lambda x: extract_value1(x, 'name'))

#genres
df1['genre'] = df1['genres'].apply(lambda x: extract_value1(x, 'name'))
df1['id_genre'] = df1['genres'].apply(lambda x: extract_value1(x, 'id'))

df1 = df1.drop('genres', axis=1)

#production_countries
df1['production_countries'] = df1['production_countries'].apply(lambda x: extract_value1(x, 'name'))

#spoken_languages
df1['spoken_languages'] = df1['spoken_languages'].apply(lambda x: extract_value1(x, 'name'))
```

```python
#Verificamos cuantos nulos hay en la columna 'release_date'con sus indices correspondientes
indices_nan = df1[df1['release_date'].isna()].index
indices_nan
```

```python
#Ahora creamos una mascara para borrar los nulos de 'release_date' y resetear su indice
mask = df['release_date'].notnull()

df1 = df1[mask]

df1 = df1.reset_index(drop=True)
```

```python
#Funcion para extraer el año de lanzamiento
def release_year(x):
    try:
        x = x.split("-")
        return x[0]
    except:
        return None
```

```python
#Aplicar la funcion y cambiar nombre por 'release_year'

df1['release_year'] = df1["release_date"].apply(lambda x: release_year(x))
```

```python
#Detectamos outliers de 'release_year'
df1['release_year'].unique()
```

```python
#Despues de detectarlos, imprimimos para ver los outliers
anio1 = df1[df1['release_year'] == '22'].index
anio1
```

```python
#Revisamos que contiene las filas de los outliers y vemos años con valores menores a 4 digitos (por ej: '22')
print(df1['release_year'].iloc[35587])
print(len(df1['release_year'].iloc[35587]))
```

```python
#Funcion para identificar outliers en release_year y borrar filas innecesarias del dataframe, sin borrar los Nonetype

mask = df1["release_year"].astype(str).str.match(r'^\d{4}$') | df1["release_year"].isna()

df1= df1[mask]

df1= df1.reset_index(drop=True) 
```

Aplicamos una mascara para borrar aquellas filas que no cumplan con la condicion de ella, que, en este caso son años de 4 digitos o que los valores sean None. Lo que hace la mascara, es recorrer la columna que damos en la primer linea, y en los casos que es false, los filtra y los elimina. (ARREGLAR)

```python
#Verificamos que la columna haya quedado limpia
df1['release_year'].unique()
```

```python
#Creamos la columna 'return' que calcula el retorno de inversión a partir de la formula 'revenue' / 'budget'.

#Primero verificamos los tipos de variables que se alojan en 'revenue' y 'budget'
type(df1["budget"].iloc[50]) #str
type(df1["revenue"].iloc[50]) #float
```

```python
#Casteamos las columnas a flotantes. 
#Luego realizamos la formula para calcular el retorno de inversion en la nueva columna 'return'
df1['budget'] = df1['budget'].astype(float)
df1['revenue'] = df1['revenue'].astype(float)

df1['return'] = np.where(
    (df1['budget'] != 0) & (df1['revenue'] != 0) & (df1['budget'].notnull()) & (df1['revenue'].notnull()),
    df1['revenue'] / df1['budget'],
    0
)

#Si alguno de los dos valores es igual a 0, la columna return se rellena con 0.
```

```python
#Verificamos la informacion
df_reducido = df1[['budget', 'revenue','return']]
df_reducido
```

```python
#Guarmados el dataframe limpia a un nuevo csv

df1.to_csv('1.movies_dataset_limpio.csv')
```
